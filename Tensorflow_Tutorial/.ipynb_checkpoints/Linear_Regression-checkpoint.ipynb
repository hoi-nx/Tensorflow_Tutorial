```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
# X=[1,2,3,4] => [x0,x1,x2,x3]
# Y=[2] =>w_init + w_0*x_0+...+w_3*x_3
#Fake data
x_data=np.random.random((10000,2))
#Fake sample
sample_weights=np.array([3,4]).reshape(2,)

#Fake y_data
y_data=np.matmul(x_data,sample_weights)
x_data[0]
```

```python
y_data[0] 
#0.4 *3+0.55*4
```

```python
#Sấp sỉ y_data để cho dữ liệu mang tính ngẫu nhiên,y_data + 1 đại lượng random trong khoảng -0.5,0.5
y_data=np.add(y_data,np.random.uniform(-0.5,0.5))
y_data[0]
```

```python
x_data.shape
```

```python
#reshape y thành ma trận 10000*1
y_data=y_data.reshape(len(y_data),1)

x_data
```

```python
y_data
```

```python
from sklearn.model_selection import train_test_split
#tách dữ liệu thành 2 phần training và testing
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
y_train.shape
```

```python
n_dim=x_train.shape[1]
x=tf.placeholder(tf.float32,[None,n_dim])
y=tf.placeholder(tf.float32,[None,1])
```

```python
#Trainning weights
W=tf.Variable(tf.ones([n_dim,1]))
b=tf.Variable(np.random.randn(),dtype=tf.float32)
```

```python
#Loss fuction
pred=tf.add(tf.matmul(x,W),b)
loss=tf.reduce_mean(tf.square(pred-y))
learning_rate=0.01
optimize=tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(loss)
```

```python
init=tf.global_variables_initializer()
```

```python
sess=tf.Session()
```

```python
#Tranning
epochs=1000
sess.run(init)
loss_history=[]

for epoch in range(epochs):
    sess.run(optimize,feed_dict={x:x_train,y:y_train})
    tess_loss=sess.run(loss,feed_dict={x:x_test,y:y_test})
    
    loss_history.append(tess_loss)
    if epochs%100==0:
        print("Epochs {} Test Lost={}".format(epoch,tess_loss))
print('Traning finished')
```

```python
sess.run(W)
sess.run(b)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(len(loss_history)),loss_history)
plt.axis([0,epochs,0,np.max(loss_history)])
plt.show()
```
