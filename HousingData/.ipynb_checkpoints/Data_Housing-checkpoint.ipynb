# Cleaning data

```python
#Gia tri tb cua total_bedrooms
median=housing['total_bedrooms'].median()
```

```python
housing['total_bedrooms'].fillna(median,inplace=True)
```

```python
housing.info()
```

# Combination feature

```python
#Tong hop cac thuoc tinh.
#So phong tb cua moi can nha
housing['room_per_household']=housing['total_rooms']/housing['households']
#So luong phong ngu tren 1 nha
housing['bedroom_per_household']=housing['total_bedrooms']/housing['households']
#So nguoi ten mot nha
housing['population_per_household']=housing['population']/housing['households']
```

```python
housing.head(4)
```

```python
#Tinh toan ma tran tuong quan
corr_matrix=housing.corr()
corr_matrix['median_house_value'].sort_values()
```

# Text to categories

```python
#Make one hot encoder
#Bien thanh chuoi ma so 0 1
housing_dummies=pd.get_dummies(housing,prefix='ocean_proximity')
```

```python
housing_dummies.head(4)
```

```python
# Scaling data
#Min max scaller: chuyen du lieu ve tu khoang min den khoan max
#Chuyen du lieu ve phan phoi chuan
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
#Bai toan dinh gia nha bo cai median_house_value
#sklearn.preprocessing.StandardScaler . chuyen du lieu theo phan phoi chuan.
housing_features=housing_dummies.drop('median_house_value',axis=1)
```

```python
housing_features.head(4)
```

```python
X_data=scaler.fit_transform(housing_features)
```

```python
X_data.shape
```

```python
y_data=np.array(housing_dummies['median_house_value'])
```

```python
y_data.shape[0]
```

```python
y_data
```

```python
y_data=y_data.reshape(y_data.shape[0],1)
```

```python
y_data.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_data,y_data,test_size=0.2,random_state=24)
```

```python
X_train.shape[1]
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#Du doan su dung mo hinh tuyen tinh
y_predict=model.predict(X_test)
```

```python
#Returns the coefficient of determination R^2 of the prediction.
model.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Sai so toan phuong trung binh,tức là sự khác biệt giữa các ước lượng và những gì được đánh giá,
#MSE là một hàm rủi ro, tương ứng với giá trị kỳ vọng của sự mất mát sai số bình phương hoặc mất mát bậc ha
def rmse_error(y_predict,y_test):
    mse_error=mean_squared_error(y_predict,y_test)
    return np.sqrt(mse_error)
```

```python
rmse_error(y_predict,y_test)
```

# Training TensorFlow

```python
#Placeholder for pass data
X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
```

```python
#Trainable weights
W=tf.Variable(tf.ones([X_train.shape[1],1]))
b=tf.Variable(np.random.randn(),dtype=tf.float32)
```

```python
pred=tf.add(tf.matmul(X,W),b)
```

```python
loss=tf.sqrt(tf.reduce_mean(tf.square(pred-Y)))
learning_rate=200
```

```python
optimize=tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(loss)
```

```python
init =tf.global_variables_initializer()
```

```python
sess=tf.Session()
```

```python
sess.run(init)
epochs=20000
loss_history=[]
for epoch in range(epochs):
    sess.run(optimize,feed_dict={X:X_train,Y:y_train})
    test_loss=sess.run(loss,feed_dict={X:X_test,Y:y_test})
    loss_history.append(test_loss)
    
    if epoch%500==0:
       print("Epoch {} Test Loss = {}".format(epoch,test_loss))
print("Training success")
```
